```python
import csv

# Load dataset from CSV file
with open('enjoy_sport.csv', 'r') as file:
    data = list(csv.reader(file))

# Separate attributes and target values
concepts = [row[:-1] for row in data[1:]]
target = [row[-1] for row in data[1:]]

print("Training Examples:")
for i in range(len(concepts)):
    print(concepts[i], "->", target[i])

# Initialize hypothesis with first positive example
hypothesis = ['0'] * len(concepts[0])

for i in range(len(concepts)):
    if target[i].lower() == 'yes':

        # If hypothesis is still empty, take first positive example
        if hypothesis == ['0'] * len(concepts[0]):
            hypothesis = concepts[i].copy()
         # Generalize hypothesis when required
        else:
            for j in range(len(hypothesis)):
                if hypothesis[j] != concepts[i][j]:
                    hypothesis[j] = '?'

        print(f"Step {i+1} Hypothesis: {hypothesis}")

print("\nFinal Hypothesis:")
print(hypothesis)    
```
