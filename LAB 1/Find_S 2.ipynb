```python
# Cell 1 : Import Library

import pandas as pd
```

```python
# Cell 2 : Read Dataset

data = pd.read_csv("tennis.csv")
data
```

```python
# Cell 3 : Show Dataset Information

data.info()
```

```python
# Cell 4 : Display First 5 Rows

data.head()
```

```python
# Cell 5 : Separate Concepts and Target

concepts = data.iloc[:, :-1].values
target = data.iloc[:, -1].values

print("Concepts:\n", concepts)
print("\nTarget:\n", target)
```

```python
# Cell 6 : Initialize Hypothesis with First Positive Example

for i in range(len(target)):
    if target[i] == "yes":
        hypothesis = concepts[i].copy()
        break

print("Initial Hypothesis:")
print(hypothesis)
```

```python
# Cell 7 : Apply Find-S Algorithm

for i in range(len(concepts)):
    
    if target[i] == "yes":
        
        for j in range(len(hypothesis)):
            
            if hypothesis[j] != concepts[i][j]:
                hypothesis[j] = '?'
        
        print("After Example", i + 1, ":", hypothesis)
```

```python
# Cell 8 : Final Hypothesis

print("\nFinal Hypothesis:")
print(hypothesis)
```
