```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

print("="*60)
print("ARTIFICIAL NEURAL NETWORK - BACKPROPAGATION")
print("="*60)

# 1. Load dataset
print("1. Loading Iris dataset...")
data = load_iris()
X = data.data
y = data.target
print("Dataset shape:", X.shape)

# One-hot encoding
y = np.eye(3)[y]

# 2. Preprocessing
print("2. Preprocessing data...")
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print("Training samples:", X_train.shape[0])
print("Test samples:", X_test.shape[0])

# 3. Create Network
print("3. Creating Neural Network...")
input_size = 4
hidden_size = 8
output_size = 3
print(f"Architecture: {input_size} -> {hidden_size} -> {output_size}")

# Initialize weights
W1 = np.random.randn(input_size, hidden_size)
W2 = np.random.randn(hidden_size, output_size)
lr = 0.1

# Functions
def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_derivative(x):
    return x*(1-x)

def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)

# 4. Training
print("4. Training Neural Network (Backpropagation)...")

for epoch in range(500):

    # Forward
    hidden = sigmoid(np.dot(X_train, W1))
    output = softmax(np.dot(hidden, W2))

    # Loss
    loss = -np.mean(np.sum(y_train * np.log(output + 1e-8), axis=1))

    # Backward
    d_output = output - y_train
    d_hidden = d_output.dot(W2.T) * sigmoid_derivative(hidden)

    # Update
    W2 -= lr * hidden.T.dot(d_output)
    W1 -= lr * X_train.T.dot(d_hidden)

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

# 5. Testing
print("5. Evaluating on test set...")

hidden_test = sigmoid(np.dot(X_test, W1))
output_test = softmax(np.dot(hidden_test, W2))

y_pred = np.argmax(output_test, axis=1)
y_true = np.argmax(y_test, axis=1)

acc = accuracy_score(y_true, y_pred)

print(f"Test Accuracy: {acc*100:.2f}%")
print("="*60)
print(f"Final Test Accuracy: {acc*100:.2f}%")
print("="*60)
```
