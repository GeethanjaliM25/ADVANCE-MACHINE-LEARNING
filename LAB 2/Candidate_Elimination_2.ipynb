```python
import pandas as pd

# Read the dataset
data = pd.read_csv("tennis.csv")

# Display dataset
print("Dataset:\n")
print(data)

# Separate concepts and target
concepts = data.iloc[:, :-1].values.tolist()
target = data.iloc[:, -1].values.tolist()

# Number of attributes
num_attributes = len(concepts[0])

# Initialize Specific Hypothesis
S = ['0'] * num_attributes

# Find first positive example
for i in range(len(target)):
    if target[i].lower() == 'yes':
        S = concepts[i].copy()
        break

# Initialize General Hypothesis
G = [['?' for i in range(num_attributes)]]

print("\nInitial Specific Hypothesis (S):", S)
print("Initial General Hypothesis (G):", G)

# Candidate Elimination Algorithm
for i in range(len(concepts)):

    if target[i].lower() == 'yes':

        # Generalize S
        for j in range(num_attributes):
            if S[j] != concepts[i][j]:
                S[j] = '?'

        # Remove inconsistent hypotheses from G
        G = [g for g in G if all(g[k] == '?' or g[k] == S[k] for k in range(num_attributes))]

    else:

        # Specialize G for negative example
        new_G = []

        for j in range(num_attributes):

            if S[j] != concepts[i][j] and S[j] != '?':
                temp = ['?' for k in range(num_attributes)]
                temp[j] = S[j]

                if temp not in new_G:
                    new_G.append(temp)

        if new_G:
            G = new_G

    print("\nExample", i + 1)
    print("Instance :", concepts[i])
    print("Target   :", target[i])
    print("S =", S)
    print("G =", G)

# Final Output
print("\nFinal Specific Hypothesis:")
print(S)

print("\nFinal General Hypothesis:")
for g in G:
    print(g)
```
